## **Job Computer Vision**


**Problem Statement :**

What are the total number of job openings related to Computer Vision ?

Note:For finding the job related to computer vision check the Job Title column.

```
Output Format

Count

```

````python
a = []
count = 0
for i in df['Title']:
  i = i.lower()
  
  if "computer" in i:
    
    count += 1
    a.append(i)
    
print(a)
print(len(a))

````

## **Job Month 2018**

Problem Statement :

Find the month having most job openings in Year 2018 ?
Print the month (Month Name i.e January, February, March) and Number of job opening as Integer Value

```
Output Format :

MonthName Count

```

````python

count = 0
l = []
d = []
count = {}


for i in data:
  b = (i['Posting_date'].split(","))
  
  #a = (i['Posting_date'].split(",")[-1].replace(" ",''))
  if '2018' in b[1]: # if 2018 in date append it to l
    l.append(b)


for i in l:
  a = i[0][0:3]
  d.append(a)

for i in d:
  if i in count:
    count[i] += 1
  else:
    count[i] = 1

print(count)
````

## **Job Degree**

Problem Statement :
Find the number of job openings are present if applicant have Bachelor degree?
Print the count as Integer value

**Note : Here we will use the BASIC QUALIFICATIONS feature to find out whether bachelor degree for Job is required or not. Keywords that can be used are 'Bachelor', 'BS' and 'BA'.**

```
Output:

Count
```

````python
a = []
for i in data:
  city = (i['BASIC QUALIFICATIONS'])
  
  if 'Bachelor' in city or 'BA' in city or 'BS' in city :
    a.append(city)

print(len(a))
````
